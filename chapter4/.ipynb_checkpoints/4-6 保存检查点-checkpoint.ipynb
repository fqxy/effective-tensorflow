```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
plotdata = { "batchsize":[], "loss":[] }
def moving_average(a, w=10):
    if len(a) < w: 
        return a[:]    
    return [val if idx < w else sum(a[(idx-w):idx])/w for idx, val in enumerate(a)]
```

```python
#生成模拟数据
train_X = np.linspace(-1, 1, 100)
train_Y = 2 * train_X + np.random.randn(*train_X.shape) * 0.3 # y=2x，但是加入了噪声
#图形显示
plt.plot(train_X, train_Y, 'ro', label='Original data')
plt.legend()
plt.show()
```

```python
tf.reset_default_graph()
```

```python
# 创建模型
# 占位符
X = tf.placeholder("float")
Y = tf.placeholder("float")
# 模型参数
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.zeros([1]), name="bias")
# 前向结构
z = tf.multiply(X, W)+ b
```

```python
#反向优化
cost =tf.reduce_mean( tf.square(Y - z))
learning_rate = 0.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost) #Gradient descent
```

```python
# 初始化变量
init = tf.global_variables_initializer()
#参数设置
training_epochs = 20
display_step = 2
```

```python
# 创建存储器saver
# max_to_keep=1表示在迭代过程中只会保存一个文件
# 这样在循环训练的过程中，新生成的模型就会覆盖以前的模型
saver = tf.train.Saver(max_to_keep=1)
savedir = "log/"
```

```python
# 启动session
with tf.Session() as sess:
    sess.run(init)

    # Fit all training data
    for epoch in range(training_epochs):
        for (x, y) in zip(train_X, train_Y):
            sess.run(optimizer, feed_dict={X: x, Y: y})

        #显示训练中的详细信息
        if epoch % display_step == 0:
            loss = sess.run(cost, feed_dict={X: train_X, Y:train_Y})
            print ("Epoch:", epoch+1, 
                   "cost=", loss,
                   "W=", sess.run(W), 
                   "b=", sess.run(b))
            if not (loss == "NA" ):
                plotdata["batchsize"].append(epoch)
                plotdata["loss"].append(loss)
                
            # 保存此时的模型
            saver.save(sess, savedir+"linermodel.cpkt", global_step=epoch)
                
    print (" Finished!")
    
    print ("cost=", sess.run(cost, feed_dict={X: train_X, Y: train_Y}), 
           "W=", sess.run(W), 
           "b=", sess.run(b))
    #print ("cost:",cost.eval({X: train_X, Y: train_Y}))

    #图形显示
    plt.plot(train_X, train_Y, 'ro', label='Original data')
    plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
    plt.legend()
    plt.show()
    
    plotdata["avgloss"] = moving_average(plotdata["loss"])
    plt.figure(1)
    plt.subplot(211)
    plt.plot(plotdata["batchsize"], plotdata["avgloss"], 'b--')
    plt.xlabel('Minibatch number')
    plt.ylabel('Loss')
    plt.title('Minibatch run vs. Training loss')
     
    plt.show()
```

```python
# 方法1   
load_epoch=18    
with tf.Session() as sess2:
    # 初始化
    sess2.run(tf.global_variables_initializer())  
    # 恢复模型
    saver.restore(sess2, savedir+"linermodel.cpkt-" + str(load_epoch))
    print ("x=0.2，z=", sess2.run(z, feed_dict={X: 0.2}))
```

```python
# 方法2
with tf.Session() as sess3:
    # 初始化
    sess3.run(tf.global_variables_initializer()) 
    # 获取检查点状态
    ckpt = tf.train.get_checkpoint_state(savedir)
    if ckpt and ckpt.model_checkpoint_path:
        # 恢复模型
        saver.restore(sess3, ckpt.model_checkpoint_path)
        print ("x=0.2，z=", sess3.run(z, feed_dict={X: 0.2}))
```

```python
# 方法3
with tf.Session() as sess4:
    # 初始化
    sess4.run(tf.global_variables_initializer())
    # 列出最近未删除的checkpoint文件名
    kpt = tf.train.latest_checkpoint(savedir)
    if kpt!=None:
        saver.restore(sess4, kpt) 
        print ("x=0.2，z=", sess4.run(z, feed_dict={X: 0.2}))
```
